```python
# Sample valid Indian cities list (could be much longer, consider using an API or database)
valid_indian_cities = [
    'Ludhiana', 'Varanasi', 'Kanpur', 'Lucknow', 'Meerut', 'Agra', 'Ghaziabad',
    'Visakhapatnam', 'Krishna', 'Mumbai', 'Thane', 'Nashik', 'Pune', 'Nagpur',
    'Kalyan', 'Vasai-Virar', 'Ahmedabad', 'Rajkot', 'Surat', 'Vadodara', 'Bangalore',
    'Patna', 'Jaipur', 'Faridabad', 'Gurgaon', 'Hyderabad', 'Srinagar', 'Kolkata',
    'Chennai', 'Bhopal', 'Indore', 'Morena', 'Delhi'
]

# Filter the dataframe for valid cities in India
valid_city_df = df[df['City'].isin(valid_indian_cities)]

# Plot the distribution of valid cities
plt.figure(figsize=(10, 6))
ax = valid_city_df['City'].value_counts().plot(kind='bar', color='skyblue')

# Add labels above the bars
for i, v in enumerate(valid_city_df['City'].value_counts()):
    ax.text(i, v + 0.2, str(v), ha='center', va='bottom', fontsize=10)

plt.title('Distribution of Cities in India (Valid Cities Only)')
plt.xlabel('City')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mapping of cities to states
city_to_state = {
    'Ludhiana': 'Punjab',
    'Varanasi': 'Uttar Pradesh', 'Kanpur': 'Uttar Pradesh', 'Lucknow': 'Uttar Pradesh',
    'Meerut': 'Uttar Pradesh', 'Agra': 'Uttar Pradesh', 'Ghaziabad': 'Uttar Pradesh',
    'Visakhapatnam': 'Andhra Pradesh', 'Krishna': 'Andhra Pradesh',
    'Mumbai': 'Maharashtra', 'Thane': 'Maharashtra', 'Nashik': 'Maharashtra', 'Pune': 'Maharashtra',
    'Nagpur': 'Maharashtra', 'Kalyan': 'Maharashtra', 'Vasai-Virar': 'Maharashtra',
    'Ahmedabad': 'Gujarat', 'Rajkot': 'Gujarat', 'Surat': 'Gujarat', 'Vadodara': 'Gujarat',
    'Bangalore': 'Karnataka',
    'Patna': 'Bihar',
    'Jaipur': 'Rajasthan',
    'Faridabad': 'Haryana', 'Gurgaon': 'Haryana',
    'Hyderabad': 'Telangana',
    'Srinagar': 'Jammu and Kashmir',
    'Kolkata': 'West Bengal',
    'Chennai': 'Tamil Nadu',
    'Bhopal': 'Madhya Pradesh', 'Indore': 'Madhya Pradesh', 'Morena': 'Madhya Pradesh',
    'Delhi': 'Delhi'
}

# Add a new column to your dataframe for the corresponding state
df['State'] = df['City'].map(city_to_state)

all_states = [
    'Andhra Pradesh', 'Arunachal Pradesh', 'Assam', 'Bihar', 'Chhattisgarh', 'Goa', 'Gujarat',
    'Haryana', 'Himachal Pradesh', 'Jharkhand', 'Karnataka', 'Kerala', 'Madhya Pradesh',
    'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha', 'Punjab', 'Rajasthan',
    'Sikkim', 'Tamil Nadu', 'Telangana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal',
    'Delhi', 'Jammu and Kashmir', 'Lakshadweep', 'Puducherry'
]
df['State'] = pd.Categorical(df['State'], categories=all_states)

# Filter the dataframe for valid cities
valid_city_df = df[df['State'].notnull()]

# Plot the distribution of states
plt.figure(figsize=(10, 6))
ax = valid_city_df['State'].value_counts().plot(kind='bar', color='lightcoral')

# Add labels above the bars
for i, v in enumerate(valid_city_df['State'].value_counts()):
    ax.text(i, v + 0.2, str(v), ha='center', va='bottom', fontsize=10)

plt.title('Distribution of Cities by State ')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Mapping of states to regions
state_to_region = {
   'Maharashtra' : 'Lower Dominated Belt (central-west)', 'Gujarat' : 'Lower Dominated Belt (central-west)',
   'Delhi'  : 'Upper Dominated Belt (central-north)', 'Uttar Pradesh'  : 'Upper Dominated Belt (central-north)',
   'Punjab'  : 'Upper Dominated Belt (central-north)', 'Haryana'  : 'Upper Dominated Belt (central-north)',
   'Bihar'  : 'Upper Dominated Belt (central-north)',
   'Madhya Pradesh' :'Other', 'West Bengal' :'Other', 'Andhra Pradesh' :'Other', 'Jammu and Kashmir' :'Other',
   'Telangana' :'Other', 'Rajasthan' :'Other', 'Karnataka' :'Other', 'Tamil Nadu' :'Other',
}

df['Region'] = df['State'].map(state_to_region)

all_regions={
     'Upper Dominated Belt (central-north)', 'Lower Dominated Belt (central-west)', 'Other', 'Not covered'
             }
df['Region'] = pd.Categorical(df['Region'], categories=all_regions)


# Filter the dataframe for valid cities and regions
valid_city_df = df[df['Region'].notnull()]

# Plot the distribution of cities by region
plt.figure(figsize=(12, 6))
ax = valid_city_df['Region'].value_counts().sort_index().plot(kind='bar', color='lightseagreen')

# Add labels above the bars
for i, v in enumerate(valid_city_df['Region'].value_counts().sort_index()):
    ax.text(i, v + 0.2, str(v), ha='center', va='bottom', fontsize=10)

custom_labels = [
    ('Other', 8),
    ('Upper Dominated Belt (central-north)', 5),
    ('Lower Dominated Belt (central-west)', 2),
    ('Not covered', 13)
]

# Get sorted region names based on value_counts().sort_index()
sorted_regions = valid_city_df['Region'].value_counts().sort_index().index

# Create a mapping of labels to their corresponding regions
label_mapping = dict(custom_labels)

for i, region in enumerate(sorted_regions):
    count = label_mapping.get(region, 0)  # Get count from custom_labels or default to 0
    bar_height = valid_city_df['Region'].value_counts().sort_index().iloc[i]

    ax.text(i, bar_height + 0.5, f"{count} states", ha='center', va='top', fontsize=10)



plt.title('Distribution of Cities by Region (8 Directions + Central India)')
plt.xlabel('Region')
plt.ylabel('Frequency')
plt.xticks(rotation=10)
plt.show()
```
