```python
#depression and gender
df_clean = df[['Gender', 'Depression']].dropna()
ct = pd.crosstab(df_clean['Gender'], df_clean['Depression'])
chi2, p, dof, ep = chi2_contingency(ct)
print('Chi-squared statistic:', chi2)
print('p-value:', p)
```
